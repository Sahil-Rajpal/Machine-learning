# one hot encoding

```python
import pandas as pd
df=pd.read_csv("homeprices2.csv")
df
```

```python
# creating dummy  variables
dummies=pd.get_dummies(df.town,dtype='int')
dummies
```

```python
# merging two data-frames:-
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
# drop original time columns:
# also drop dummy variables column to avoid dummy variables trap (if we have n dummy variables we will remove 1 variables to keep n-1 dummy variables)
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price',axis='columns') # to remove the price column as it is dependent variable
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```
