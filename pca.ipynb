```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
x=df
y=dataset.target
```

```python
x
```

```python
y
```

```python
# SCALING THE FEATURES BEFORE APPLYTING PCA
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scale=scaler.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95) # 0.95 represents that it will retain 95% of importanyt features
x_pca=pca.fit_transform(x)
x_pca.shape
```

```python
x.shape
```

```python
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
x_train_pca,x_test_pca,y_train_pca,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=30)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train_pca,y_train_pca)
```

```python
model.score(x_test_pca,y_test)
```
